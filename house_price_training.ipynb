```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("house_data.csv")
df
```

Find median of bedrooms to fill the missing NaN


```python
import math
median_Bedrooms = math.floor(df.Bedrooms.median())
median_Bedrooms
```

Fill in the missing value in the Column

```python
df.Bedrooms.fillna(median_Bedrooms)
```

Final data frame after filling the missing value

```python
df.Bedrooms = df.Bedrooms.fillna(median_Bedrooms)
df
```

##Training model

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Area','Bedrooms','Age']], df.Price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

Predicting price of home with the following inputs

```python
reg.predict([[ 3000, 3, 40 ]])
```
